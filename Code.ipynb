PART G : 
Combine the datasets from 2015-2019 to construct 1 dataframe that
contains the following columns. Country, Region, Year (coming from
the respective dataset) and Happiness Score for each year as a new
column.



```python
df=pd.read_csv('2015.csv')
df1=pd.read_csv('2016.csv')
df2=pd.read_csv('2017.csv')
df3=pd.read_csv('2018.csv')
df4=pd.read_csv('2019.csv')
```

```python
#Extract the required Coloumns
df = df[['Country', 'Happiness Score']]
df['Year'] = 2015

df1 = df1[['Country', 'Happiness Score']]
df1['Year'] = 2016

df2 = df2[['Country', 'Happiness.Score']]
df2.columns = ['Country', 'Happiness Score']  
df2['Year'] = 2017

df3 = df3[['Country or region', 'Score']]
df3.columns = ['Country', 'Happiness Score'] 
df3['Year'] = 2018

df4 = df4[['Country or region', 'Score']]
df4.columns = ['Country', 'Happiness Score'] 
df4['Year'] = 2019

#Concatination of dataframes
combined = pd.concat([df, df1, df2, df3, df4], ignore_index=True)

#Have Years as a Coloumm
dataframe_new = combined.pivot_table(index=['Country'], columns='Year', values='Happiness Score').reset_index()


dataframe_new.to_csv('combinedData.csv', index=False)

# Print the combined DataFrame
print(dataframe_new)
```

PART J :  Identify which 2 variables have the highest correlations across all the
countries for the years 2015 and 2016.



```python
df=pd.read_csv('2015.csv')
df1=pd.read_csv('2016.csv')

columnskeep = [
    'Happiness Score',
    'Economy (GDP per Capita)',
    'Family',
    'Health (Life Expectancy)',
    'Freedom',
    'Trust (Government Corruption)',
    'Generosity',
    'Dystopia Residual'
]

df = df[columnskeep]
df1 = df1[columnskeep]

# Calculate the correlation matrix for each year
corel15 = df.corr()
corel16 = df1.corr()

# Find the pair of variables with the highest correlation for each year
paircorel15 = corel15.unstack().sort_values(ascending=False).drop_duplicates().head(2)
paircorel16 = corel16.unstack().sort_values(ascending=False).drop_duplicates().head(2)

# Extract variable names (column names) and their correlation values for highest correlation pairs
highestpaircorel15 = [(pair[0], pair[1], corel15[pair[0]][pair[1]]) for pair in paircorel15.index]
highestpaircorel16 = [(pair[0], pair[1], corel16[pair[0]][pair[1]]) for pair in paircorel16.index]

print("The Variables having highest correlation for the Year 2015 :\n ")
for x, y, z in highestpaircorel15:
    print(f"{x} and {y}: {z:.2f}")

print('\n')
print("The Variables having highest correlation for the Year 2016 :\n ")
for x, y, z in highestpaircorel16:
    print(f"{x} and {y}: {z:.2f}")
```

PART A : 
Identify and rectify any missing values in the data using appropriate
techniques.


```python
import pandas as pd

df = pd.read_csv('earth_surface_temperatures.csv')

valuesmissing = df.isnull().sum()
print(valuesmissing)
df.dropna(inplace=True)
```

```python
df['Temperature'].fillna(df['Temperature'].mean(), inplace=True)
```

```python
df.interpolate(inplace=True)
```

```python
valuesmissing1 = df.isnull().sum()
print(valuesmissing1)
```

PART B : 
Ensure that the data types of all columns are consistent with their
values and make conversions where necessary.


```python
print(df.dtypes)
```

```python
df['Years'] = df['Years'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Country'] = df['Country'].astype(str)
df['Temperature'] = df['Temperature'].astype(float)
df['Monthly_variation'] = df['Monthly_variation'].astype(float)
df['Anomaly'] = df['Anomaly'].astype(float)
```

```python
print(df.dtypes)
```

PART C : Transform the Years and Month columns into a single column
labeled &quot;Date&quot; in the MM-YYYY format, with a datetime64[ns] data
type. For example, the year 1848 and month 5 should be unified as a
single value, such as 5-1848.


```python
df['Date'] = df['Month'].astype(str) + '-' + df['Years'].astype(str)

#add new coloumn
df['Date'] = pd.to_datetime(df['Date'], format='%m-%Y')

#drop previous coloumn
df.drop(['Years', 'Month'], axis=1, inplace=True)

# Display the updated DataFrame
print(df.head())
```

PART D : Detect and investigate extreme temperature values that might be
regarded as outliers.


```python
Q1 = df['Temperature'].quantile(0.25)
Q3 = df['Temperature'].quantile(0.75)

IQR = Q3 - Q1

lowerbound = Q1 - 1.5 * IQR
upperbound = Q3 + 1.5 * IQR

outliers = df[(df['Temperature'] < lowerbound) | (df['Temperature'] > upperbound)]

print("Lower Bound (LB):", lowerbound)
print('\n')
print("Upper Bound (UB):", upperbound)
print('\n')
print("Interquartile Range (IQR):", IQR)
print('\n')
print("Outliers:")
print(outliers)
print('\n')
```

PART G : Determine the overall trend in global temperatures over the years and
visualize this trend using a suitable chart.

```python
import matplotlib.pyplot as plt


df = pd.read_csv('earth_surface_temperatures.csv')
avg = df.groupby('Years')['Temperature'].mean()

# Plot the trend using a line plot
plt.figure(figsize=(8, 5))
plt.plot(avg.index, avg.values, marker='v', color='red')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.title('The Trend of Mean temperature over The Years')
plt.grid(False)
plt.show()
```

PART J :  Choose five countries and compare the trends in their temperatures
over the years, seeking any similar temperature patterns.


```python
countries = ['Afghanistan', 'El Salvador', 'Equatorial Guinea', 'Eritrea', 'Isle of Man']
filtereddata = df[df['Country'].isin(countries)]

groupeddata = filtereddata.groupby(['Country', 'Years'])['Temperature'].mean().unstack()

plt.figure(figsize=(15, 6))

for country in groupeddata:
    plt.plot(groupeddata.index, groupeddata[country], label=country)

plt.xlabel('Countries')
plt.ylabel('Temperature')
plt.title('The Trend of Temperature Over Years\n')
plt.legend()
plt.grid(True)
plt.show()
```

PART K:  Explore the potential correlation between temperature and monthly
variation or anomaly values. Calculate correlation coefficients and
create scatterplots to investigate this relationship.

```python
#Calculate Corelations
tempvar = df['Temperature'].corr(df['Monthly_variation'])
tempana = df['Temperature'].corr(df['Anomaly'])

plt.figure(figsize=(10, 6))

#Create Scatter
plt.subplot(1, 2, 1)
plt.scatter(df['Temperature'], df['Monthly_variation'], alpha=0.7, color='red')
plt.title(f'Temperature and Monthly Variation ({tempvar:.2f})')
plt.xlabel('Temperature')
plt.ylabel('Monthly Variation')

plt.subplot(1, 2, 2)
plt.scatter(df['Temperature'], df['Anomaly'], alpha=0.7, color='red')
plt.title(f'Temperature and Anamoly ({tempana:.2f})')
plt.xlabel('Temperature')
plt.ylabel('Anomaly')

plt.tight_layout()
plt.show()
```
